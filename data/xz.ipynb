```python
"""
Nephrotic Syndrome Genetic Diversity Database Setup - UPDATED VERSION
This script creates and populates a SQLite database with enhanced links and annotations
"""

import sqlite3
import pandas as pd
import os
from pathlib import Path

# Database Schema Design - UPDATED
DATABASE_SCHEMA = """
-- Table 1: Genes - Core gene information with enhanced annotations
CREATE TABLE IF NOT EXISTS genes (
    gene_id INTEGER PRIMARY KEY AUTOINCREMENT,
    gene_name TEXT UNIQUE NOT NULL,
    ensembl_id TEXT,
    chromosome TEXT,
    start_position INTEGER,
    end_position INTEGER,
    srns_ssns TEXT,
    phenotypes TEXT,
    inheritance TEXT,
    pediatric TEXT,
    adult TEXT,
    both TEXT,
    gene_function TEXT,
    pathway TEXT,
    category TEXT,
    variation TEXT,
    created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);

-- Table 2: Gene Links - All database and literature links
CREATE TABLE IF NOT EXISTS gene_links (
    link_id INTEGER PRIMARY KEY AUTOINCREMENT,
    gene_name TEXT NOT NULL,
    omim_link TEXT,
    gwas_link TEXT,
    clinpgx_link TEXT,
    malacards_link TEXT,
    function_link TEXT,
    pathway_link TEXT,
    source_paper_1 TEXT,
    source_paper_2 TEXT,
    source_paper_3 TEXT,
    FOREIGN KEY (gene_name) REFERENCES genes(gene_name)
);

-- Table 3: Variants - All variant information with dbSNP links
CREATE TABLE IF NOT EXISTS variants (
    variant_id INTEGER PRIMARY KEY AUTOINCREMENT,
    rs_id TEXT UNIQUE NOT NULL,
    chrom TEXT NOT NULL,
    position INTEGER NOT NULL,
    ref_allele TEXT NOT NULL,
    alt_allele TEXT NOT NULL,
    gene_name TEXT NOT NULL,
    global_af REAL,
    dbsnp_link TEXT,
    FOREIGN KEY (gene_name) REFERENCES genes(gene_name),
    UNIQUE(chrom, position, ref_allele, alt_allele)
);

-- Table 4: Allele Frequencies - Population-specific frequencies
CREATE TABLE IF NOT EXISTS allele_frequencies (
    af_id INTEGER PRIMARY KEY AUTOINCREMENT,
    variant_id INTEGER NOT NULL,
    population TEXT NOT NULL,
    allele_frequency REAL NOT NULL,
    FOREIGN KEY (variant_id) REFERENCES variants(variant_id),
    UNIQUE(variant_id, population)
);

-- Table 5: Statistical Tests - FST and Chi-square results
CREATE TABLE IF NOT EXISTS statistical_tests (
    test_id INTEGER PRIMARY KEY AUTOINCREMENT,
    variant_id INTEGER NOT NULL,
    population TEXT NOT NULL,
    fst_value REAL,
    chi_square_p_value REAL,
    is_significant_chisq BOOLEAN DEFAULT 0,
    is_significant_bonferroni BOOLEAN DEFAULT 0,
    is_significant_fst BOOLEAN DEFAULT 0,
    FOREIGN KEY (variant_id) REFERENCES variants(variant_id),
    UNIQUE(variant_id, population)
);

-- Table 6: LD Status - Linkage Disequilibrium information
CREATE TABLE IF NOT EXISTS ld_status (
    ld_id INTEGER PRIMARY KEY AUTOINCREMENT,
    variant_id INTEGER NOT NULL,
    population TEXT NOT NULL,
    in_ld BOOLEAN DEFAULT 0,
    ld_pruned BOOLEAN DEFAULT 0,
    FOREIGN KEY (variant_id) REFERENCES variants(variant_id),
    UNIQUE(variant_id, population)
);

-- Indexes for faster queries
CREATE INDEX IF NOT EXISTS idx_variants_gene ON variants(gene_name);
CREATE INDEX IF NOT EXISTS idx_variants_rs ON variants(rs_id);
CREATE INDEX IF NOT EXISTS idx_variants_chrom_pos ON variants(chrom, position);
CREATE INDEX IF NOT EXISTS idx_af_variant ON allele_frequencies(variant_id);
CREATE INDEX IF NOT EXISTS idx_af_population ON allele_frequencies(population);
CREATE INDEX IF NOT EXISTS idx_stats_variant ON statistical_tests(variant_id);
CREATE INDEX IF NOT EXISTS idx_ld_variant ON ld_status(variant_id);
CREATE INDEX IF NOT EXISTS idx_gene_links ON gene_links(gene_name);
"""

class NephroticSyndromeDB:
    def __init__(self, db_path='nephrotic_syndrome.db'):
        """Initialize database connection"""
        self.db_path = db_path
        self.conn = None
        self.cursor = None
        
    def connect(self):
        """Connect to SQLite database"""
        self.conn = sqlite3.connect(self.db_path)
        self.cursor = self.conn.cursor()
        print(f"✓ Connected to database: {self.db_path}")
        
    def create_schema(self):
        """Create database schema"""
        self.cursor.executescript(DATABASE_SCHEMA)
        self.conn.commit()
        print("✓ Database schema created successfully")
        
    def load_gene_data(self, excel_path):
        """Load gene information from Excel file - UPDATED"""
        print("\n📊 Loading gene data...")
        df = pd.read_excel(excel_path)
        
        # Clean column names
        df.columns = df.columns.str.strip()
        
        gene_count = 0
        for idx, row in df.iterrows():
            try:
                # Insert gene information
                self.cursor.execute("""
                    INSERT OR IGNORE INTO genes (
                        gene_name, ensembl_id, chromosome, start_position, end_position,
                        srns_ssns, phenotypes, inheritance, pediatric, adult, both,
                        gene_function, pathway, category, variation
                    ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
                """, (
                    row.get('Gene'),
                    row.get('Ensembl_ID'),
                    row.get('chromosome'),
                    row.get('start'),
                    row.get('end'),
                    row.get('SRNS/SSNS'),
                    row.get('Phenotypes'),
                    row.get('Inheritance'),
                    row.get('Pediatric (Y/N)') or row.get(' Pediatric (Y/N)'),
                    row.get('Adult (Y/N)'),
                    row.get('Both (Y/N)'),
                    row.get('Function'),
                    row.get('Pathway'),
                    row.get('Category'),
                    row.get('Variation')
                ))
                
                # Insert link information
                self.cursor.execute("""
                    INSERT OR IGNORE INTO gene_links (
                        gene_name, omim_link, gwas_link, clinpgx_link, malacards_link,
                        function_link, pathway_link,
                        source_paper_1, source_paper_2, source_paper_3
                    ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
                """, (
                    row.get('Gene'),
                    row.get('OMIM_Link'),
                    row.get('GWAS_Link'),
                    row.get('ClinPGx_Link'),
                    row.get('MalaCards_Link'),
                    row.get('Function'),  # This is the function link
                    row.get('Pathway'),   # This is the pathway link
                    row.get('Source Paper 1'),
                    row.get('Source Paper 2'),
                    row.get('Source Paper 3')
                ))
                
                gene_count += 1
                
            except Exception as e:
                print(f"Error inserting gene {row.get('Gene')}: {e}")
                
        self.conn.commit()
        print(f"✓ Loaded {gene_count} genes")
        
    def load_variant_data(self, tsv_path):
        """Load variant and statistical test data from main TSV file - UPDATED"""
        print("\n📊 Loading variant data (this may take a few minutes)...")
        
        # Read in chunks to handle large file
        chunk_size = 10000
        chunks_processed = 0
        
        for chunk in pd.read_csv(tsv_path, sep='\t', chunksize=chunk_size):
            variant_count = 0
            
            for idx, row in chunk.iterrows():
                try:
                    # Insert variant with dbSNP link
                    self.cursor.execute("""
                        INSERT OR IGNORE INTO variants (
                            rs_id, chrom, position, ref_allele, alt_allele, 
                            gene_name, global_af, dbsnp_link
                        ) VALUES (?, ?, ?, ?, ?, ?, ?, ?)
                    """, (
                        row['ID'],
                        row['CHROM'],
                        row['POS'],
                        row['REF'],
                        row['ALT'],
                        row['GENE'],
                        row['AF'],
                        row.get('db_SNP_link')
                    ))
                    
                    # Get variant_id
                    variant_id = self.cursor.execute(
                        "SELECT variant_id FROM variants WHERE rs_id = ?",
                        (row['ID'],)
                    ).fetchone()[0]
                    
                    # Insert allele frequencies for all populations
                    populations = {
                        'AFR': row['AF_AFR'], 'AMR': row['AF_AMR'], 'AMR1': row['AF_AMR1'],
                        'AMR2': row['AF_AMR2'], 'EAS': row['AF_EAS'], 'EUR': row['AF_EUR'],
                        'SAS': row['AF_SAS'], 'BEB': row['AF_BEB'], 'GIH': row['AF_GIH'],
                        'ITU': row['AF_ITU'], 'PJL': row['AF_PJL'], 'STU': row['AF_STU']
                    }
                    
                    for pop, freq in populations.items():
                        self.cursor.execute("""
                            INSERT OR IGNORE INTO allele_frequencies (variant_id, population, allele_frequency)
                            VALUES (?, ?, ?)
                        """, (variant_id, pop, freq))
                    
                    # Insert statistical test results
                    stat_populations = ['AFR', 'AMR', 'AMR1', 'AMR2', 'EAS', 'SAS', 'BEB', 'GIH', 'ITU', 'PJL', 'STU']
                    
                    for pop in stat_populations:
                        fst_col = f'FST_EUR_VS_{pop}'
                        p_col = f'P_EUR_{pop}'
                        
                        self.cursor.execute("""
                            INSERT OR IGNORE INTO statistical_tests (
                                variant_id, population, fst_value, chi_square_p_value
                            ) VALUES (?, ?, ?, ?)
                        """, (variant_id, pop, row[fst_col], row[p_col]))
                    
                    variant_count += 1
                    
                except Exception as e:
                    print(f"Error processing variant {row.get('ID')}: {e}")
                    continue
            
            chunks_processed += 1
            self.conn.commit()
            print(f"  Processed chunk {chunks_processed} ({variant_count} variants)")
            
        print(f"✓ Loaded all variants")
        
    def load_significant_variants(self, folder_path, ld_status='before'):
        """Load significant variants and mark them in the database"""
        print(f"\n📊 Loading significant variants ({ld_status} LD)...")
        
        test_types = {
            'Significant Chisq result': 'chisq',
            'Significant_Chisq_result': 'chisq',
            'Significant_Chisq_result_Bonferroni': 'bonferroni',
            'Significant_Bonferroni_with_FST': 'bonferroni',
            'Significant_Fst': 'fst'
        }
        
        populations = ['AFR', 'AMR', 'AMR1', 'AMR2', 'EAS', 'SAS', 'BEB', 'GIH', 'ITU', 'PJL', 'STU']
        
        total_updated = 0
        total_skipped = 0
        
        for test_folder, test_type in test_types.items():
            test_path = os.path.join(folder_path, test_folder)
            if not os.path.exists(test_path):
                continue
                
            for pop in populations:
                # Find matching file
                files = list(Path(test_path).glob(f"{pop}_*.tsv"))
                if not files:
                    continue
                    
                file_path = files[0]
                
                try:
                    df = pd.read_csv(file_path, sep='\t')
                    
                    for _, row in df.iterrows():
                        rs_id = row['ID']
                        
                        # Check if variant exists in database
                        result = self.cursor.execute(
                            "SELECT variant_id FROM variants WHERE rs_id = ?",
                            (rs_id,)
                        ).fetchone()
                        
                        if result is None:
                            total_skipped += 1
                            continue
                        
                        variant_id = result[0]
                        
                        # Update significance flags
                        update_query = f"""
                            UPDATE statistical_tests
                            SET is_significant_{test_type} = 1
                            WHERE variant_id = ? AND population = ?
                        """
                        self.cursor.execute(update_query, (variant_id, pop))
                        
                        # Update LD status
                        self.cursor.execute("""
                            INSERT OR REPLACE INTO ld_status (variant_id, population, in_ld, ld_pruned)
                            VALUES (?, ?, ?, ?)
                        """, (variant_id, pop, 1 if ld_status == 'after' else 0, ld_status == 'after'))
                        
                        total_updated += 1
                        
                except Exception as e:
                    print(f"  Warning: Error processing {file_path.name}: {e}")
                    
        self.conn.commit()
        print(f"✓ Loaded significant variants ({ld_status} LD)")
        print(f"  Updated: {total_updated} variant-population combinations")
        if total_skipped > 0:
            print(f"  Skipped: {total_skipped} variants not found in main dataset")
        
    def close(self):
        """Close database connection"""
        if self.conn:
            self.conn.close()
            print("\n✓ Database connection closed")


# Main execution function
def main():
    print("=" * 60)
    print("Nephrotic Syndrome Genetics Database Setup - UPDATED")
    print("=" * 60)
    
    # Initialize database
    db = NephroticSyndromeDB('nephrotic_syndrome.db')
    db.connect()
    
    # Create schema
    db.create_schema()
    
    # Load data (Update these paths to your actual file locations)
    print("\n" + "=" * 60)
    print("LOADING DATA")
    print("=" * 60)
    
    # 1. Load gene information
    db.load_gene_data('Final_gene_list.xlsx')
    
    # 2. Load variant data
    db.load_variant_data('Final_with_dbsnp.tsv')
    
    # 3. Load significant variants (Before LD)
    db.load_significant_variants('Before_LD', ld_status='before')
    
    # 4. Load significant variants (After LD)
    db.load_significant_variants('After_LD', ld_status='after')
    
    # Close connection
    db.close()
    
    print("\n" + "=" * 60)
    print("DATABASE SETUP COMPLETE!")
    print("=" * 60)
    print(f"\nYour database 'nephrotic_syndrome.db' is ready to use!")
    print("\nNext steps:")
    print("1. Test queries with the query examples script")
    print("2. Set up your web dashboard using Flask/Django")
    print("3. Deploy your application")


if __name__ == "__main__":
    main()
```

```python
"""
Nephrotic Syndrome Genetic Diversity Database Setup - CORRECTED VERSION
Fixed to match actual Excel column names - Function and Pathway are LINKS
"""

import sqlite3
import pandas as pd
import os
from pathlib import Path

# ==================== SIGNIFICANCE CRITERIA ====================
CHISQ_P_THRESHOLD = 0.05        # Chi-square p-value threshold (p <= 0.05)
CHISQ_AF_THRESHOLD = 0.05       # Minimum allele frequency (AF >= 0.05 in EITHER population)
BONFERRONI_P_THRESHOLD = 0.01   # Bonferroni corrected p-value threshold (p <= 0.01)
FST_THRESHOLD = 0.15            # FST threshold for significant differentiation (FST >= 0.15)

# ==================== DATABASE SCHEMA ====================
DATABASE_SCHEMA = """
-- Table 1: Genes - Core gene information
CREATE TABLE IF NOT EXISTS genes (
    gene_id INTEGER PRIMARY KEY AUTOINCREMENT,
    gene_name TEXT UNIQUE NOT NULL,
    ensembl_id TEXT,
    chromosome TEXT,
    start_position INTEGER,
    end_position INTEGER,
    srns_ssns TEXT,
    phenotypes TEXT,
    inheritance TEXT,
    pediatric TEXT,
    adult TEXT,
    both TEXT,
    category TEXT,
    variation TEXT,
    created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);

-- Table 2: Gene Links - All database and literature links (including Function and Pathway links)
CREATE TABLE IF NOT EXISTS gene_links (
    link_id INTEGER PRIMARY KEY AUTOINCREMENT,
    gene_name TEXT NOT NULL,
    omim_link TEXT,
    gwas_link TEXT,
    clinpgx_link TEXT,
    malacards_link TEXT,
    function_link TEXT,
    pathway_link TEXT,
    source_paper_1 TEXT,
    source_paper_2 TEXT,
    source_paper_3 TEXT,
    FOREIGN KEY (gene_name) REFERENCES genes(gene_name),
    UNIQUE(gene_name)
);

-- Table 3: Variants - All variant information with dbSNP links
CREATE TABLE IF NOT EXISTS variants (
    variant_id INTEGER PRIMARY KEY AUTOINCREMENT,
    rs_id TEXT UNIQUE NOT NULL,
    chrom TEXT NOT NULL,
    position INTEGER NOT NULL,
    ref_allele TEXT NOT NULL,
    alt_allele TEXT NOT NULL,
    gene_name TEXT NOT NULL,
    global_af REAL,
    dbsnp_link TEXT,
    FOREIGN KEY (gene_name) REFERENCES genes(gene_name),
    UNIQUE(chrom, position, ref_allele, alt_allele)
);

-- Table 4: Allele Frequencies - Population-specific frequencies
CREATE TABLE IF NOT EXISTS allele_frequencies (
    af_id INTEGER PRIMARY KEY AUTOINCREMENT,
    variant_id INTEGER NOT NULL,
    population TEXT NOT NULL,
    allele_frequency REAL NOT NULL,
    FOREIGN KEY (variant_id) REFERENCES variants(variant_id),
    UNIQUE(variant_id, population)
);

-- Table 5: Statistical Tests - FST and Chi-square results with significance flags
CREATE TABLE IF NOT EXISTS statistical_tests (
    test_id INTEGER PRIMARY KEY AUTOINCREMENT,
    variant_id INTEGER NOT NULL,
    population TEXT NOT NULL,
    fst_value REAL,
    chi_square_p_value REAL,
    is_significant_chisq BOOLEAN DEFAULT 0,
    is_significant_bonferroni BOOLEAN DEFAULT 0,
    is_significant_fst BOOLEAN DEFAULT 0,
    FOREIGN KEY (variant_id) REFERENCES variants(variant_id),
    UNIQUE(variant_id, population)
);

-- Table 6: LD Status - Linkage Disequilibrium information
CREATE TABLE IF NOT EXISTS ld_status (
    ld_id INTEGER PRIMARY KEY AUTOINCREMENT,
    variant_id INTEGER NOT NULL,
    population TEXT DEFAULT 'ALL',
    in_ld BOOLEAN DEFAULT 0,
    ld_pruned BOOLEAN DEFAULT 0,
    FOREIGN KEY (variant_id) REFERENCES variants(variant_id),
    UNIQUE(variant_id, population)
);

-- Indexes for faster queries
CREATE INDEX IF NOT EXISTS idx_variants_gene ON variants(gene_name);
CREATE INDEX IF NOT EXISTS idx_variants_rs ON variants(rs_id);
CREATE INDEX IF NOT EXISTS idx_variants_chrom_pos ON variants(chrom, position);
CREATE INDEX IF NOT EXISTS idx_af_variant ON allele_frequencies(variant_id);
CREATE INDEX IF NOT EXISTS idx_af_population ON allele_frequencies(population);
CREATE INDEX IF NOT EXISTS idx_stats_variant ON statistical_tests(variant_id);
CREATE INDEX IF NOT EXISTS idx_stats_significant ON statistical_tests(is_significant_bonferroni);
CREATE INDEX IF NOT EXISTS idx_ld_variant ON ld_status(variant_id);
CREATE INDEX IF NOT EXISTS idx_ld_pruned ON ld_status(ld_pruned);
CREATE INDEX IF NOT EXISTS idx_gene_links ON gene_links(gene_name);
CREATE INDEX IF NOT EXISTS idx_genes_chrom ON genes(chromosome);
"""


class NephroticSyndromeDB:
    def __init__(self, db_path='nephrotic_syndrome.db'):
        """Initialize database connection"""
        self.db_path = db_path
        self.conn = None
        self.cursor = None
        
    def connect(self):
        """Connect to SQLite database"""
        self.conn = sqlite3.connect(self.db_path)
        self.cursor = self.conn.cursor()
        print(f"✓ Connected to database: {self.db_path}")
        
    def create_schema(self):
        """Create database schema"""
        print("\nCreating database schema...")
        self.cursor.executescript(DATABASE_SCHEMA)
        self.conn.commit()
        print("✓ Database schema created successfully")
        
    def load_gene_data(self, excel_path):
        """Load gene information from Excel file"""
        print("\n" + "="*70)
        print("LOADING GENE DATA")
        print("="*70)
        print(f"Reading from: {excel_path}")
        
        df = pd.read_excel(excel_path)
        
        # Clean column names (remove extra spaces)
        df.columns = df.columns.str.strip()
        
        print(f"Found {len(df)} genes in Excel file")
        print(f"\nColumns in Excel: {list(df.columns)}")
        
        gene_count = 0
        link_count = 0
        
        for idx, row in df.iterrows():
            try:
                gene_name = row.get('Gene')
                
                if pd.isna(gene_name):
                    continue
                
                # Insert gene information (removed gene_function and pathway as they are links)
                self.cursor.execute("""
                    INSERT OR IGNORE INTO genes (
                        gene_name, ensembl_id, chromosome, start_position, end_position,
                        srns_ssns, phenotypes, inheritance,
                        pediatric, adult, both, category, variation
                    ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
                """, (
                    gene_name,
                    row.get('Ensembl_ID'),
                    row.get('chromosome'),
                    row.get('start'),
                    row.get('end'),
                    row.get('SRNS/SSNS'),
                    row.get('Phenotypes'),
                    row.get('Inheritance'),
                    row.get('Pediatric (Y/N)'),
                    row.get('Adult (Y/N)'),
                    row.get('Both (Y/N)'),
                    row.get('Category'),
                    row.get('Variation')
                ))
                
                gene_count += 1
                
                # Insert gene links - Function and Pathway are LINKS, not text
                self.cursor.execute("""
                    INSERT OR IGNORE INTO gene_links (
                        gene_name, omim_link, gwas_link, clinpgx_link, malacards_link,
                        function_link, pathway_link,
                        source_paper_1, source_paper_2, source_paper_3
                    ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
                """, (
                    gene_name,
                    row.get('OMIM_Link'),
                    row.get('GWAS_Link'),
                    row.get('ClinPGx_Link'),
                    row.get('MalaCards_Link'),
                    row.get('Function'),      # This is a LINK
                    row.get('Pathway'),       # This is a LINK
                    row.get('Source Paper 1'),
                    row.get('Source Paper 2'),
                    row.get('Source Paper 3')
                ))
                
                link_count += 1
                
            except Exception as e:
                print(f"  ⚠️  Error inserting gene {row.get('Gene')}: {e}")
                
        self.conn.commit()
        print(f"✓ Loaded {gene_count} genes")
        print(f"✓ Loaded {link_count} gene link records (including function & pathway links)")
        
    def load_variant_data(self, tsv_path):
        """Load variant and statistical test data from main TSV file"""
        print("\n" + "="*70)
        print("LOADING VARIANT DATA")
        print("="*70)
        print(f"Reading from: {tsv_path}")
        print("\nThis may take several minutes for large files...")
        
        # Define populations
        populations = ['AFR', 'AMR', 'AMR1', 'AMR2', 'EAS', 'SAS', 'BEB', 'GIH', 'ITU', 'PJL', 'STU']
        
        # Read in chunks to handle large files
        chunk_size = 10000
        total_variants = 0
        chunks_processed = 0
        
        print(f"\n📊 Significance Criteria:")
        print(f"  • Chi-square: p ≤ {CHISQ_P_THRESHOLD} AND AF ≥ {CHISQ_AF_THRESHOLD} (in EITHER population)")
        print(f"  • Bonferroni: p ≤ {BONFERRONI_P_THRESHOLD}")
        print(f"  • FST: FST ≥ {FST_THRESHOLD}\n")
        
        for chunk in pd.read_csv(tsv_path, sep='\t', chunksize=chunk_size, low_memory=False):
            for idx, row in chunk.iterrows():
                try:
                    # Create dbSNP link
                    dbsnp_link = row.get('db_SNP_link')
                    if pd.isna(dbsnp_link) or not dbsnp_link:
                        dbsnp_link = f"https://www.ncbi.nlm.nih.gov/snp/{row['ID']}"
                    
                    # Insert variant with dbSNP link
                    self.cursor.execute("""
                        INSERT OR IGNORE INTO variants (
                            rs_id, chrom, position, ref_allele, alt_allele, 
                            gene_name, global_af, dbsnp_link
                        ) VALUES (?, ?, ?, ?, ?, ?, ?, ?)
                    """, (
                        row['ID'],
                        row['CHROM'],
                        row['POS'],
                        row['REF'],
                        row['ALT'],
                        row['GENE'],
                        row.get('AF', 0),
                        dbsnp_link
                    ))
                    
                    # Get variant_id
                    result = self.cursor.execute(
                        "SELECT variant_id FROM variants WHERE rs_id = ?",
                        (row['ID'],)
                    ).fetchone()
                    
                    if not result:
                        continue
                    
                    variant_id = result[0]
                    
                    # Get EUR allele frequency for significance calculations
                    af_eur = row.get('AF_EUR', 0)
                    
                    # Insert allele frequencies for EUR
                    self.cursor.execute("""
                        INSERT OR IGNORE INTO allele_frequencies (variant_id, population, allele_frequency)
                        VALUES (?, ?, ?)
                    """, (variant_id, 'EUR', af_eur))
                    
                    # Insert allele frequencies for all other populations
                    for pop in populations:
                        af_col = f'AF_{pop}'
                        if af_col in row.index:
                            af_pop = row[af_col]
                            if not pd.isna(af_pop):
                                self.cursor.execute("""
                                    INSERT OR IGNORE INTO allele_frequencies (variant_id, population, allele_frequency)
                                    VALUES (?, ?, ?)
                                """, (variant_id, pop, af_pop))
                    
                    # Insert statistical tests with significance flags
                    for pop in populations:
                        fst_col = f'FST_EUR_VS_{pop}'
                        p_col = f'P_EUR_{pop}'
                        af_pop_col = f'AF_{pop}'
                        
                        if fst_col in row.index and p_col in row.index:
                            fst_value = row[fst_col]
                            p_value = row[p_col]
                            af_pop = row.get(af_pop_col, 0)
                            
                            # Skip if values are NaN
                            if pd.isna(fst_value) or pd.isna(p_value):
                                continue
                            
                            # Calculate significance flags
                            af_sufficient = (af_eur >= CHISQ_AF_THRESHOLD) or (af_pop >= CHISQ_AF_THRESHOLD)
                            is_sig_chisq = (p_value <= CHISQ_P_THRESHOLD) and af_sufficient
                            is_sig_bonf = p_value <= BONFERRONI_P_THRESHOLD
                            is_sig_fst = fst_value >= FST_THRESHOLD
                            
                            self.cursor.execute("""
                                INSERT OR IGNORE INTO statistical_tests (
                                    variant_id, population, fst_value, chi_square_p_value,
                                    is_significant_chisq, is_significant_bonferroni, is_significant_fst
                                ) VALUES (?, ?, ?, ?, ?, ?, ?)
                            """, (
                                variant_id, pop, fst_value, p_value,
                                1 if is_sig_chisq else 0,
                                1 if is_sig_bonf else 0,
                                1 if is_sig_fst else 0
                            ))
                    
                    # Initialize LD status as NOT pruned
                    self.cursor.execute("""
                        INSERT OR IGNORE INTO ld_status (variant_id, population, in_ld, ld_pruned)
                        VALUES (?, ?, ?, ?)
                    """, (variant_id, 'ALL', 0, 0))
                    
                    total_variants += 1
                    
                except Exception as e:
                    print(f"  ⚠️  Error processing variant {row.get('ID')}: {e}")
                    continue
            
            chunks_processed += 1
            
            # Commit every chunk
            self.conn.commit()
            
            if chunks_processed % 10 == 0:
                print(f"  Processed {total_variants:,} variants...")
        
        print(f"\n✓ Loaded {total_variants:,} variants with statistical tests")
        
    def load_ld_pruned_variants(self, after_ld_dir):
        """Mark variants that PASSED LD pruning from After_LD folder"""
        print("\n" + "="*70)
        print("MARKING LD-PRUNED VARIANTS")
        print("="*70)
        print(f"Scanning directory: {after_ld_dir}")
        
        after_path = Path(after_ld_dir)
        
        if not after_path.exists():
            print(f"⚠️  Warning: After_LD directory not found at {after_path}")
            print("All variants will be marked as NOT LD-pruned")
            return
        
        ld_passed_variants = set()
        files_processed = 0
        
        print("\nScanning all TSV files...")
        
        for file in after_path.rglob('*.tsv'):
            try:
                df = pd.read_csv(file, sep='\t', low_memory=False)
                
                if 'ID' in df.columns:
                    variant_ids = df['ID'].unique()
                    ld_passed_variants.update(variant_ids)
                    files_processed += 1
                    
                    if files_processed % 10 == 0:
                        print(f"  Processed {files_processed} files, {len(ld_passed_variants):,} unique variants found...")
                        
            except Exception as e:
                print(f"  ⚠️  Error reading {file.name}: {e}")
        
        print(f"\n📁 Files processed: {files_processed}")
        print(f"🧬 Unique variants found (PASSED LD): {len(ld_passed_variants):,}")
        
        if len(ld_passed_variants) == 0:
            print("⚠️  Warning: No variants found in After_LD files!")
            return
        
        print("\nUpdating database...")
        
        # First, ensure all variants have ld_pruned = 0
        self.cursor.execute("UPDATE ld_status SET ld_pruned = 0, in_ld = 0")
        
        # Mark the ones that passed as ld_pruned = 1
        updated_count = 0
        batch_size = 1000
        
        ld_variants_list = list(ld_passed_variants)
        for i in range(0, len(ld_variants_list), batch_size):
            batch = ld_variants_list[i:i+batch_size]
            
            placeholders = ','.join(['?'] * len(batch))
            self.cursor.execute(f'''
                UPDATE ld_status 
                SET ld_pruned = 1, in_ld = 1
                WHERE variant_id IN (
                    SELECT variant_id FROM variants WHERE rs_id IN ({placeholders})
                )
            ''', batch)
            
            updated_count += len(batch)
            
            if updated_count % 5000 == 0:
                print(f"  Updated {updated_count:,} / {len(ld_passed_variants):,} variants...")
                self.conn.commit()
        
        self.conn.commit()
        
        # Show statistics
        self.cursor.execute('''
            SELECT 
                SUM(CASE WHEN ld_pruned = 1 THEN 1 ELSE 0 END) as kept,
                SUM(CASE WHEN ld_pruned = 0 THEN 1 ELSE 0 END) as removed,
                COUNT(*) as total
            FROM ld_status
        ''')
        stats = self.cursor.fetchone()
        
        print(f"\n📊 LD Pruning Statistics:")
        print(f"  • Variants KEPT (passed LD): {stats[0]:,}")
        print(f"  • Variants REMOVED (pruned): {stats[1]:,}")
        print(f"  • Total variants: {stats[2]:,}")
        
        if stats[2] > 0:
            retention_rate = (stats[0]/stats[2]*100)
            print(f"  • Retention rate: {retention_rate:.2f}%")
    
    def update_gene_positions(self):
        """Update gene chromosome and position info from variants (if not in Excel)"""
        print("\n" + "="*70)
        print("UPDATING GENE POSITIONS")
        print("="*70)
        
        # Only update genes that don't have positions from Excel
        self.cursor.execute('''
            UPDATE genes
            SET chromosome = (
                SELECT DISTINCT chrom
                FROM variants
                WHERE variants.gene_name = genes.gene_name
                LIMIT 1
            ),
            start_position = (
                SELECT MIN(position)
                FROM variants
                WHERE variants.gene_name = genes.gene_name
            ),
            end_position = (
                SELECT MAX(position)
                FROM variants
                WHERE variants.gene_name = genes.gene_name
            )
            WHERE (chromosome IS NULL OR start_position IS NULL OR end_position IS NULL)
            AND EXISTS (
                SELECT 1 FROM variants WHERE variants.gene_name = genes.gene_name
            )
        ''')
        
        updated = self.cursor.rowcount
        self.conn.commit()
        print(f"✓ Updated positions for {updated} genes from variant data")
    
    def show_statistics(self):
        """Display database statistics"""
        print("\n" + "="*70)
        print("DATABASE STATISTICS")
        print("="*70)
        
        # Genes
        self.cursor.execute("SELECT COUNT(*) FROM genes")
        total_genes = self.cursor.fetchone()[0]
        print(f"\n📚 Total Genes: {total_genes:,}")
        
        # Variants
        self.cursor.execute("SELECT COUNT(*) FROM variants")
        total_variants = self.cursor.fetchone()[0]
        print(f"🧬 Total Variants: {total_variants:,}")
        
        # Significant variants (Bonferroni)
        self.cursor.execute('''
            SELECT COUNT(DISTINCT variant_id)
            FROM statistical_tests
            WHERE is_significant_bonferroni = 1
        ''')
        sig_variants = self.cursor.fetchone()[0]
        if total_variants > 0:
            print(f"✨ Significant Variants (Bonferroni): {sig_variants:,} ({sig_variants/total_variants*100:.2f}%)")
        
        # LD statistics
        self.cursor.execute('''
            SELECT 
                SUM(CASE WHEN ld_pruned = 1 THEN 1 ELSE 0 END) as kept,
                SUM(CASE WHEN ld_pruned = 0 THEN 1 ELSE 0 END) as pruned
            FROM ld_status
        ''')
        ld_stats = self.cursor.fetchone()
        if ld_stats:
            print(f"\n🔗 LD Pruning:")
            print(f"  • Kept: {ld_stats[0]:,}")
            print(f"  • Pruned: {ld_stats[1]:,}")
        
        # Significance by population
        self.cursor.execute('''
            SELECT population, COUNT(DISTINCT variant_id) as count
            FROM statistical_tests
            WHERE is_significant_bonferroni = 1
            GROUP BY population
            ORDER BY count DESC
            LIMIT 10
        ''')
        
        print(f"\n🌍 Top Populations with Significant Variants (Bonferroni):")
        for row in self.cursor.fetchall():
            print(f"  • {row[0]}: {row[1]:,}")
        
        print("\n" + "="*70)
    
    def close(self):
        """Close database connection"""
        if self.conn:
            self.conn.close()
            print("\n✓ Database connection closed")


def main():
    print("="*70)
    print("NEPHROTIC SYNDROME GENETICS DATABASE - CORRECTED VERSION")
    print("="*70)
    print("\n📏 Using Significance Criteria:")
    print(f"  • Chi-square: p ≤ {CHISQ_P_THRESHOLD} AND AF ≥ {CHISQ_AF_THRESHOLD} (in EITHER population)")
    print(f"  • Bonferroni: p ≤ {BONFERRONI_P_THRESHOLD}")
    print(f"  • FST: FST ≥ {FST_THRESHOLD}")
    
    # Initialize database
    db = NephroticSyndromeDB('nephrotic_syndrome.db')
    db.connect()
    
    # Create schema
    db.create_schema()
    
    # ==================== LOAD DATA ====================
    # Update these paths to match your file locations
    
    # 1. Load gene information from Excel
    db.load_gene_data('Final_gene_list.xlsx')
    
    # 2. Load variant data from main TSV (with dbSNP links)
    db.load_variant_data('Final_with_dbsnp.tsv')
    
    # 3. Mark LD-pruned variants from After_LD folder
    db.load_ld_pruned_variants('After_LD/')
    
    # 4. Update gene positions from variant data (only if missing)
    db.update_gene_positions()
    
    # 5. Show statistics
    db.show_statistics()
    
    # Close connection
    db.close()
    
    print("\n" + "="*70)
    print("✅ DATABASE SETUP COMPLETE!")
    print("="*70)
    print(f"\nDatabase file: nephrotic_syndrome.db")
    print("\nNext steps:")
    print("1. Start Flask API: python app.py")
    print("2. Open dashboard in browser")
    print("3. Test queries and explore your data")


if __name__ == "__main__":
    main()
```

```python
"""
Multi-Population Significant Variants Analysis
Find variants that are significant in 2 or more populations
"""

import sqlite3
import pandas as pd
from tabulate import tabulate

DATABASE = 'nephrotic_syndrome.db'

def get_connection():
    return sqlite3.connect(DATABASE)

def query1_multi_pop_summary():
    """Variants significant in multiple populations - Summary"""
    print("\n" + "="*100)
    print("VARIANTS SIGNIFICANT IN MULTIPLE POPULATIONS - SUMMARY")
    print("="*100)
    print("Criteria: p ≤ 0.01 AND AF ≥ 0.05 (either pop) AND FST ≥ 0.15")
    print("="*100 + "\n")
    
    query = """
    SELECT 
        v.rs_id,
        v.gene_name,
        v.chrom || ':' || v.position AS position,
        COUNT(DISTINCT st.population) AS population_count,
        GROUP_CONCAT(DISTINCT st.population) AS populations,
        ROUND(AVG(st.chi_square_p_value), 8) AS avg_p_value,
        ROUND(AVG(st.fst_value), 4) AS avg_fst
    FROM statistical_tests st
    INNER JOIN variants v ON st.variant_id = v.variant_id
    INNER JOIN allele_frequencies af_eur 
        ON st.variant_id = af_eur.variant_id 
        AND af_eur.population = 'EUR'
    INNER JOIN allele_frequencies af_pop 
        ON st.variant_id = af_pop.variant_id 
        AND af_pop.population = st.population
    WHERE 
        st.chi_square_p_value <= 0.01
        AND (af_eur.allele_frequency >= 0.05 OR af_pop.allele_frequency >= 0.05)
        AND st.fst_value >= 0.15
    GROUP BY v.rs_id, v.gene_name, v.chrom, v.position
    HAVING COUNT(DISTINCT st.population) >= 2
    ORDER BY COUNT(DISTINCT st.population) DESC, avg_fst DESC
    LIMIT 50
    """
    
    conn = get_connection()
    df = pd.read_sql_query(query, conn)
    conn.close()
    
    if len(df) == 0:
        print("⚠️  No variants found that are significant in 2+ populations!")
        print("This could mean:")
        print("  • The criteria are very strict (which is good for quality)")
        print("  • Most variants show population-specific effects")
        return None
    
    # Truncate long population lists
    df['populations'] = df['populations'].apply(
        lambda x: x if len(str(x)) < 60 else str(x)[:57] + '...'
    )
    
    print(tabulate(df, headers='keys', tablefmt='pretty', showindex=False))
    print(f"\n✓ Found {len(df)} variants significant in 2+ populations")
    print(f"✓ Top variant: {df.iloc[0]['rs_id']} in {df.iloc[0]['population_count']} populations")
    
    return df

def query2_distribution():
    """Distribution of variants by population count"""
    print("\n" + "="*100)
    print("DISTRIBUTION: HOW MANY POPULATIONS ARE VARIANTS SIGNIFICANT IN?")
    print("="*100 + "\n")
    
    query = """
    WITH variant_pop_counts AS (
        SELECT 
            v.rs_id,
            COUNT(DISTINCT st.population) AS pop_count
        FROM statistical_tests st
        INNER JOIN variants v ON st.variant_id = v.variant_id
        INNER JOIN allele_frequencies af_eur 
            ON st.variant_id = af_eur.variant_id 
            AND af_eur.population = 'EUR'
        INNER JOIN allele_frequencies af_pop 
            ON st.variant_id = af_pop.variant_id 
            AND af_pop.population = st.population
        WHERE 
            st.chi_square_p_value <= 0.01
            AND (af_eur.allele_frequency >= 0.05 OR af_pop.allele_frequency >= 0.05)
            AND st.fst_value >= 0.15
        GROUP BY v.rs_id
    )
    SELECT 
        pop_count AS 'Number_of_Populations',
        COUNT(*) AS 'Variant_Count',
        ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM variant_pop_counts), 2) AS 'Percentage'
    FROM variant_pop_counts
    GROUP BY pop_count
    ORDER BY pop_count DESC
    """
    
    conn = get_connection()
    df = pd.read_sql_query(query, conn)
    conn.close()
    
    print(tabulate(df, headers='keys', tablefmt='pretty', showindex=False))
    
    total = df['Variant_Count'].sum()
    multi_pop = df[df['Number_of_Populations'] >= 2]['Variant_Count'].sum()
    
    print(f"\n📊 Summary:")
    print(f"  • Total significant variants: {total:,}")
    print(f"  • Multi-population (2+): {multi_pop:,} ({multi_pop/total*100:.1f}%)")
    print(f"  • Single population only: {total-multi_pop:,} ({(total-multi_pop)/total*100:.1f}%)")
    
    return df

def query3_genes_with_multi_pop_variants():
    """Genes with most multi-population significant variants"""
    print("\n" + "="*100)
    print("GENES WITH MOST MULTI-POPULATION SIGNIFICANT VARIANTS")
    print("="*100 + "\n")
    
    query = """
    WITH multi_pop_variants AS (
        SELECT 
            v.gene_name,
            v.rs_id,
            COUNT(DISTINCT st.population) AS pop_count
        FROM statistical_tests st
        INNER JOIN variants v ON st.variant_id = v.variant_id
        INNER JOIN allele_frequencies af_eur 
            ON st.variant_id = af_eur.variant_id 
            AND af_eur.population = 'EUR'
        INNER JOIN allele_frequencies af_pop 
            ON st.variant_id = af_pop.variant_id 
            AND af_pop.population = st.population
        WHERE 
            st.chi_square_p_value <= 0.01
            AND (af_eur.allele_frequency >= 0.05 OR af_pop.allele_frequency >= 0.05)
            AND st.fst_value >= 0.15
        GROUP BY v.gene_name, v.rs_id
        HAVING COUNT(DISTINCT st.population) >= 2
    )
    SELECT 
        gene_name AS 'Gene',
        COUNT(*) AS 'Multi_Pop_Variants',
        ROUND(AVG(pop_count), 2) AS 'Avg_Pops_Per_Variant',
        MAX(pop_count) AS 'Max_Populations',
        SUM(pop_count) AS 'Total_Pop_Hits'
    FROM multi_pop_variants
    GROUP BY gene_name
    ORDER BY COUNT(*) DESC, AVG(pop_count) DESC
    LIMIT 20
    """
    
    conn = get_connection()
    df = pd.read_sql_query(query, conn)
    conn.close()
    
    if len(df) == 0:
        print("⚠️  No genes found with multi-population variants")
        return None
    
    print(tabulate(df, headers='keys', tablefmt='pretty', showindex=False))
    print(f"\n✓ {len(df)} genes have variants significant in 2+ populations")
    
    return df

def query4_specific_variant(rs_id):
    """Detailed analysis for a specific variant across all populations"""
    print("\n" + "="*100)
    print(f"DETAILED ANALYSIS FOR {rs_id} - ALL POPULATIONS")
    print("="*100 + "\n")
    
    query = """
    SELECT 
        st.population,
        ROUND(st.chi_square_p_value, 10) AS p_value,
        ROUND(st.fst_value, 4) AS fst,
        ROUND(af_eur.allele_frequency, 4) AS af_eur,
        ROUND(af_pop.allele_frequency, 4) AS af_pop,
        CASE WHEN st.chi_square_p_value <= 0.01 THEN '✓' ELSE '✗' END AS p_check,
        CASE WHEN (af_eur.allele_frequency >= 0.05 OR af_pop.allele_frequency >= 0.05) 
             THEN '✓' ELSE '✗' END AS af_check,
        CASE WHEN st.fst_value >= 0.15 THEN '✓' ELSE '✗' END AS fst_check,
        CASE 
            WHEN st.chi_square_p_value <= 0.01 
             AND (af_eur.allele_frequency >= 0.05 OR af_pop.allele_frequency >= 0.05)
             AND st.fst_value >= 0.15 
            THEN '✅ PASS'
            ELSE '❌ FAIL'
        END AS result
    FROM statistical_tests st
    INNER JOIN variants v ON st.variant_id = v.variant_id
    INNER JOIN allele_frequencies af_eur 
        ON st.variant_id = af_eur.variant_id 
        AND af_eur.population = 'EUR'
    INNER JOIN allele_frequencies af_pop 
        ON st.variant_id = af_pop.variant_id 
        AND af_pop.population = st.population
    WHERE v.rs_id = ?
    ORDER BY 
        CASE 
            WHEN st.chi_square_p_value <= 0.01 
             AND (af_eur.allele_frequency >= 0.05 OR af_pop.allele_frequency >= 0.05)
             AND st.fst_value >= 0.15 
            THEN 1 ELSE 2 
        END,
        st.fst_value DESC
    """
    
    conn = get_connection()
    df = pd.read_sql_query(query, conn, params=(rs_id,))
    conn.close()
    
    if len(df) == 0:
        print(f"⚠️  Variant {rs_id} not found in database")
        return None
    
    print(tabulate(df, headers='keys', tablefmt='pretty', showindex=False))
    
    passed = df[df['result'] == '✅ PASS']
    print(f"\n📊 Summary for {rs_id}:")
    print(f"  • Populations tested: {len(df)}")
    print(f"  • Populations passing ALL THREE criteria: {len(passed)}")
    
    if len(passed) > 0:
        print(f"  • Passing populations: {', '.join(passed['population'].tolist())}")
    
    return df

def query5_sas_analysis():
    """Variants significant across South Asian populations"""
    print("\n" + "="*100)
    print("VARIANTS SIGNIFICANT IN MULTIPLE SOUTH ASIAN POPULATIONS")
    print("="*100)
    print("Looking at: BEB, GIH, ITU, PJL, STU")
    print("="*100 + "\n")
    
    query = """
    WITH sas_significant AS (
        SELECT 
            v.rs_id,
            v.gene_name,
            st.population,
            st.fst_value
        FROM statistical_tests st
        INNER JOIN variants v ON st.variant_id = v.variant_id
        INNER JOIN allele_frequencies af_eur 
            ON st.variant_id = af_eur.variant_id 
            AND af_eur.population = 'EUR'
        INNER JOIN allele_frequencies af_pop 
            ON st.variant_id = af_pop.variant_id 
            AND af_pop.population = st.population
        WHERE 
            st.population IN ('BEB', 'GIH', 'ITU', 'PJL', 'STU')
            AND st.chi_square_p_value <= 0.01
            AND (af_eur.allele_frequency >= 0.05 OR af_pop.allele_frequency >= 0.05)
            AND st.fst_value >= 0.15
    )
    SELECT 
        rs_id,
        gene_name,
        COUNT(DISTINCT population) AS sas_pop_count,
        GROUP_CONCAT(population) AS sas_populations,
        ROUND(AVG(fst_value), 4) AS avg_fst
    FROM sas_significant
    GROUP BY rs_id, gene_name
    HAVING COUNT(DISTINCT population) >= 2
    ORDER BY COUNT(DISTINCT population) DESC, avg_fst DESC
    """
    
    conn = get_connection()
    df = pd.read_sql_query(query, conn)
    conn.close()
    
    if len(df) == 0:
        print("⚠️  No variants found significant in 2+ South Asian populations")
        return None
    
    print(tabulate(df, headers='keys', tablefmt='pretty', showindex=False))
    print(f"\n✓ {len(df)} variants significant in 2+ South Asian populations")
    
    return df

def export_multi_pop_variants(min_populations=2, filename=None):
    """Export multi-population variants to CSV"""
    print("\n" + "="*100)
    print(f"EXPORTING VARIANTS SIGNIFICANT IN {min_populations}+ POPULATIONS")
    print("="*100 + "\n")
    
    query = """
    WITH multi_pop_counts AS (
        SELECT 
            v.variant_id,
            v.rs_id,
            COUNT(DISTINCT st.population) AS pop_count
        FROM statistical_tests st
        INNER JOIN variants v ON st.variant_id = v.variant_id
        INNER JOIN allele_frequencies af_eur 
            ON st.variant_id = af_eur.variant_id 
            AND af_eur.population = 'EUR'
        INNER JOIN allele_frequencies af_pop 
            ON st.variant_id = af_pop.variant_id 
            AND af_pop.population = st.population
        WHERE 
            st.chi_square_p_value <= 0.01
            AND (af_eur.allele_frequency >= 0.05 OR af_pop.allele_frequency >= 0.05)
            AND st.fst_value >= 0.15
        GROUP BY v.variant_id, v.rs_id
        HAVING COUNT(DISTINCT st.population) >= ?
    )
    SELECT 
        v.rs_id,
        v.gene_name,
        v.chrom,
        v.position,
        v.ref_allele,
        v.alt_allele,
        mpc.pop_count AS total_significant_populations,
        st.population,
        ROUND(st.chi_square_p_value, 10) AS p_value,
        ROUND(st.fst_value, 6) AS fst,
        ROUND(af_eur.allele_frequency, 6) AS af_eur,
        ROUND(af_pop.allele_frequency, 6) AS af_population,
        CASE WHEN ld.ld_pruned = 1 THEN 'Yes' ELSE 'No' END AS passed_ld_pruning,
        v.dbsnp_link
    FROM multi_pop_counts mpc
    INNER JOIN variants v ON mpc.variant_id = v.variant_id
    INNER JOIN statistical_tests st ON v.variant_id = st.variant_id
    INNER JOIN allele_frequencies af_eur 
        ON st.variant_id = af_eur.variant_id 
        AND af_eur.population = 'EUR'
    INNER JOIN allele_frequencies af_pop 
        ON st.variant_id = af_pop.variant_id 
        AND af_pop.population = st.population
    LEFT JOIN ld_status ld ON v.variant_id = ld.variant_id
    WHERE 
        st.chi_square_p_value <= 0.01
        AND (af_eur.allele_frequency >= 0.05 OR af_pop.allele_frequency >= 0.05)
        AND st.fst_value >= 0.15
    ORDER BY mpc.pop_count DESC, v.rs_id, st.fst_value DESC
    """
    
    conn = get_connection()
    df = pd.read_sql_query(query, conn, params=(min_populations,))
    conn.close()
    
    if len(df) == 0:
        print(f"⚠️  No variants found significant in {min_populations}+ populations")
        return None
    
    if filename is None:
        filename = f'multi_population_variants_{min_populations}plus.csv'
    
    df.to_csv(filename, index=False)
    
    unique_variants = df['rs_id'].nunique()
    print(f"✓ Exported {len(df)} rows ({unique_variants} unique variants)")
    print(f"✓ Saved to: {filename}")
    
    return df

def main():
    """Run all analyses"""
    print("\n" + "🧬"*50)
    print("MULTI-POPULATION SIGNIFICANT VARIANTS ANALYSIS")
    print("🧬"*50)
    
    # Query 1: Summary
    df1 = query1_multi_pop_summary()
    
    # Query 2: Distribution
    df2 = query2_distribution()
    
    # Query 3: Genes
    df3 = query3_genes_with_multi_pop_variants()
    
    # Query 4: Specific variant example (if exists)
    if df1 is not None and len(df1) > 0:
        example_variant = df1.iloc[0]['rs_id']
        df4 = query4_specific_variant(example_variant)
    
    # Query 5: South Asian analysis
    df5 = query5_sas_analysis()
    
    # Export option
    print("\n" + "="*100)
    print("EXPORT OPTIONS")
    print("="*100)
    
    export_choice = input("\nExport multi-population variants to CSV? (yes/no): ").lower().strip()
    
    if export_choice == 'yes':
        min_pops = input("Minimum number of populations (default=2): ").strip()
        min_pops = int(min_pops) if min_pops.isdigit() else 2
        export_multi_pop_variants(min_populations=min_pops)
    
    print("\n" + "="*100)
    print("✅ ANALYSIS COMPLETE!")
    print("="*100)
    
    # Summary
    if df1 is not None:
        print(f"\n📊 Key Findings:")
        print(f"  • Variants in 2+ populations: {len(df1['rs_id'].unique())}")
        if len(df1) > 0:
            print(f"  • Top variant: {df1.iloc[0]['rs_id']} ({df1.iloc[0]['gene_name']}) in {df1.iloc[0]['population_count']} populations")

if __name__ == "__main__":
    try:
        main()
    except Exception as e:
        print(f"\n❌ Error: {e}")
        print("\nMake sure:")
        print("1. Database file 'nephrotic_syndrome.db' exists")
        print("2. Install required packages: pip install pandas tabulate")
```
